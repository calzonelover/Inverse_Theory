```python
import numpy as np
x = np.fromfile(file='x2.dat',dtype='float32',count=20*20)
y = np.fromfile(file='y2.dat',dtype='float32',count=39*20)
x = x.reshape(20,20)
y = y.reshape(20,39)
print(x.shape, y.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(32, input_shape=(39,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(20, activation='tanh'))
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
model.fit(y, x, epochs=1000, batch_size=20)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
a = y[10]
a = a.reshape(1, 39)
b = model.predict(a)
# a = np.fromfile(file='a.dat',dtype='float32',count=20)
# a = a.reshape(1,20)
# b = model.predict(a)
plt.plot(x[10].reshape(20,), label="True Sharp")
plt.plot(a.reshape(39,), label="Smooth")
plt.plot(b.reshape(20,), label="Model Sharp")
plt.legend()
```

```python
a = np.zeros(20)
a[10] = -1.0
a = a.reshape(1,20)
b = model.predict(a)
plt.plot(a.reshape(20,))
plt.plot(b.reshape(39,))
```

```python
import numpy as np
x = np.fromfile(file='x3.dat',dtype='float32',count=20*40)
y = np.fromfile(file='y3.dat',dtype='float32',count=39*40)
x = x.reshape(40,20)
y = y.reshape(40,39)
print(x.shape, y.shape)
```

```python
plt.plot(x[30,:])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(20, input_shape=(20,), activation='linear'))
model.add(Dense(39, activation='linear'))
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
model.fit(x, y, epochs=1000, batch_size=20)
```

```python
a = np.zeros(20)
a[10] = -1.5
a[5] = 1.0
a[15] = 0.5
a = a.reshape(1,20)
b = model.predict(a)
plt.plot(a.reshape(20,))
plt.plot(b.reshape(39,))
```
