```python
import numpy as np
x = np.fromfile(file='x2.dat',dtype='float32',count=20*20)
y = np.fromfile(file='y2.dat',dtype='float32',count=39*20)
x = x.reshape(20,20,1) # (N_DAT, N_X, N_CHANNEL)
y = y.reshape(20,39)
print(x.shape, y.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Dense, ZeroPadding1D, Flatten
model = Sequential()
filter_len = 20
# model.add(ZeroPadding1D(10, ))
model.add(Conv1D(filters=8, kernel_size=filter_len, activation='relu', padding='same', input_shape=(20,1)))
model.add(MaxPooling1D(pool_size=4, strides=2))
model.add(Conv1D(filters=4, kernel_size=10, activation='relu', padding='same'))
# model.add(MaxPooling1D(pool_size=4, strides=2))
# model.add(Conv1D(filters=4, kernel_size=4, activation='relu', padding='same'))
# model.add(MaxPooling1D(pool_size=2, strides=2))
model.add(Flatten())
model.add(Dense(8, activation='relu'))
model.add(Dense(39, activation='linear'))
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=1200)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
i = 5
print(x.shape, y.shape)
plt.plot(x[i,:,0])
plt.plot(y[i,:])
```

```python
a = np.fromfile(file='a.dat',dtype='float32',count=20)
a = a.reshape(1,20,1)
b = model.predict(a)
plt.plot(a.reshape(20,), label="real_input")
plt.plot(b.reshape(39,), label="predict_smooth")
plt.legend()
```
