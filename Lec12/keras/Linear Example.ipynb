```python
import numpy as np
N_SAMPLE = 100
MATRIX_SIZE = 10

A = np.random.rand(MATRIX_SIZE, MATRIX_SIZE)
x = np.random.rand(N_SAMPLE, MATRIX_SIZE, 1)
b = np.matmul(A, x).reshape(N_SAMPLE, MATRIX_SIZE)
x = x.reshape(N_SAMPLE, MATRIX_SIZE)
print(A.shape, x.shape, b.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Dense, ZeroPadding1D, Flatten

model = Sequential()
model.add(Dense(MATRIX_SIZE, activation='linear', use_bias=False, input_shape=(MATRIX_SIZE,)))
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
model.fit(x, b, epochs=1200)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
i = 5
print(x.shape, b.shape)
plt.plot(x[i], label="x")
plt.plot(b[i], label="real_b")
plt.legend()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
i = 5
c = model.predict(x[i].reshape(1, 10))
plt.plot(x[i], label="x")
plt.plot(b[i], label="real_b")
plt.plot(c[0], label="pred_b")
plt.legend()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

model_A = model.layers[0].get_weights()[0]

fig, axs = plt.subplots(1, 3)
# fig.suptitle('Vertically stacked subplots')
axs[0].set_title('Real A')
im0 = axs[0].imshow(A, label="A")
plt.colorbar(im0, ax=axs[0])
axs[1].set_title('Model A')
im1 = axs[1].imshow(model_A, label="model_A")
plt.colorbar(im1, ax=axs[1])
axs[2].set_title('Diff')
im2 = axs[2].imshow(np.abs(np.subtract(A, model_A)))
plt.colorbar(im2, ax=axs[2])
# axs[2].colorbar()
```

# Inverse model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Dense, ZeroPadding1D, Flatten

model = Sequential()
model.add(Dense(MATRIX_SIZE, activation='linear', use_bias=False, input_shape=(MATRIX_SIZE,)))
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
history = model.fit(b, x, epochs=5000, batch_size=64)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.subplot(121);plt.plot(history.history['loss']);plt.title('Loss');
plt.subplot(122);plt.plot(history.history['accuracy']);plt.title('Accuracy');
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

inv_A = np.linalg.inv(A)
model_inv_A = model.layers[0].get_weights()[0]

fig, axs = plt.subplots(1, 3)
axs[0].set_title('Real inv A')
im0 = axs[0].imshow(inv_A, label="A")
plt.colorbar(im0, ax=axs[0])
axs[1].set_title('Model inv A')
im1 = axs[1].imshow(model_inv_A, label="model_A")
plt.colorbar(im1, ax=axs[1])
axs[2].set_title('Diff')
im2 = axs[2].imshow(np.abs(np.subtract(inv_A, model_inv_A)))
plt.colorbar(im2, ax=axs[2])
```
