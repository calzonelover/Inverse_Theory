```python
import matplotlib.pyplot as p
import numpy as n
x = n.linspace(0,1,100)
p.plot(x,x,label="linear")
p.plot(x,x**2,label="quadratic")
p.plot(x,x**3,label="cubic")
p.legend()
p.xlabel("x")
p.ylabel("y")
p.title("simple plot")
```

```python
x = n.arange(0,1,0.01)
y = n.sin(2*n.pi*x)
p.plot(x,y)
```

```python
d1,d2,d3 = n.random.rand(3,100)
f, (a1,a2) = p.subplots(1,2)
a1.plot(d1,d2,'-x',markersize=10,linewidth=4)
a2.plot(d1,d3,color="red",marker='o',linestyle='dashed')
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(1, 1), dpi=80, facecolor='w', edgecolor='k')
p.plot(d1,d2)
```

```python
def f(t):
    return n.exp(-t) * n.cos(2*n.pi*t)

t1 = n.arange(0.0, 5.0, 0.1)
t2 = n.arange(0.0, 5.0, 0.02)

                #figsize=(width,height)
figure(num=None, figsize=(8, 3), dpi=100, facecolor='w', edgecolor='k')
p.figure(1)
p.subplot(121)
p.plot(t1, f(t1), 'bo', t2, f(t2), 'k', markersize=4)

# text and title with LaTeX
p.text(2, .5, r'$\mu_1^2=100,\ \sigma=15$')
p.title(r'$\sigma_i=15$')

# annotation
p.annotate('annotation', xy=(1, .35), xytext=(2, .75),
             arrowprops=dict(facecolor='black', width=1, headwidth=7, headlength=15))

p.subplot(122)
p.plot(t2, n.cos(2*n.pi*t2), 'r--')
p.show()
```
